```python
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from dtreeviz.trees import dtreeviz
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn import tree
from sklearn.datasets import fetch_openml, load_boston, load_iris
from sklearn.metrics import (accuracy_score, confusion_matrix, f1_score,
                             mean_squared_error, precision_score, recall_score,
                             roc_auc_score)
from sklearn.model_selection import train_test_split
from sklearn.utils.multiclass import unique_labels
```

```python
def calc_mse(y_true, y_pred, title=None):
  print('----------------------------------------')
  if title is None:
    print(' MSE')
  else:
    print(title)
  print('----------------------------------------')

  print('MSE : {:.2f}'.format(mean_squared_error(y_true, y_pred)))
```

```python
def calc_clf_metrics(y_true, y_pred, title=None):
  print('----------------------------------------')
  if title is None:
    print(' Metrics')
  else:
    print(title)
  print('----------------------------------------')

  if y_true.unique().shape[0] == 2:
    print('AUC      : {:.2f}%'.format(roc_auc_score(y_true, y_pred)*100))
  
  if y_true.unique().shape[0] == 2:
    mode = 'binary'
  else:
    mode = 'macro'

  print('Accuracy : {:.2f}%'.format(accuracy_score(y_true, y_pred)*100))
  print('Recall    : {:.2f}%'.format(recall_score(y_true, y_pred, average=mode)*100))
  print('Precision : {:.2f}%'.format(precision_score(y_true, y_pred, average=mode)*100))
  print('F1       : {:.2f}%'.format(f1_score(y_true, y_pred, average=mode)*100))
```

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    """
    Refer to: https://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html

    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if not title:
        if normalize:
            title = 'Normalized confusion matrix'
        else:
            title = 'Confusion matrix, without normalization'

    # Compute confusion matrix
    cm = confusion_matrix(y_true, y_pred)
    # Only use the labels that appear in the data
    classes = classes[unique_labels(y_true, y_pred)]
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    fig, ax = plt.subplots(figsize=(10, 10))
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, fontsize=25)
    plt.yticks(tick_marks, fontsize=25)
    plt.xlabel('Predicted label',fontsize=25)
    plt.ylabel('True label', fontsize=25)
    plt.title(title, fontsize=30)

    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size="5%", pad=0.15)
    cbar = ax.figure.colorbar(im, ax=ax, cax=cax)
    cbar.ax.tick_params(labelsize=20)

    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
#            title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    fontsize=20,
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

## 2値分類

```python
titanic = fetch_openml(data_id=40945, as_frame=True)
df = pd.concat([titanic.data, titanic.target], axis=1)  # print(df.shape) >> (1309, 14)

# 不要列の削除
df = df[df.columns.difference(['boat', 'body', 'home.dest', 'name', 'cabin', 'ticket'])]
# 欠損値の補完
df['age'] = df['age'].fillna(df['age'].mean())
df['fare'] = df['fare'].fillna(df['fare'].mean())
df["embarked"] = df["embarked"].fillna("S")

# データ型変換
df = df.astype({'embarked':object, 'sex':object, 'survived':object})  # print(df.shape) >> (1309, 9)

# 特徴量と目的変数の分離
X = df[df.columns.difference(['survived'])]
y = df['survived'].astype(int)

# カテゴリ変数のエンコード
X['sex'].replace({'male':0, 'female':1}, inplace=True)
X = pd.get_dummies(X, columns=['embarked'])

print(X.shape, y.shape)
display(X.head())
display(y.head())

# train/test分割 (train:test = 7:3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  shuffle=True, random_state=100)
print('学習データ : ', X_train.shape, y_train.shape)
print('テストデータ : ', X_test.shape, y_test.shape)
```

```python
# モデル学習
clf = tree.DecisionTreeClassifier(max_depth=5, min_samples_split=10, random_state=100)
clf = clf.fit(X_train, y_train)

# 予測値算出
train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test) 
print(train_pred.shape, test_pred.shape)

# 評価指標算出
calc_clf_metrics(y_train, train_pred, title='Metrics (train)')
calc_clf_metrics(y_test, test_pred, title='Metrics (test)')

# Confusion matrix作成
class_names = np.array(['dead', 'survive'])
print('\n')
plot_confusion_matrix(y_train, train_pred, class_names, normalize=True, title='Confusion matrix (train)')
print('\n')
plot_confusion_matrix(y_test, test_pred, class_names, normalize=True, title='Confusion matrix (test)')
```

```python
# 決定木の描画
viz = dtreeviz(
    clf,
    X_train,
    y_train,
    target_name='survived',
    feature_names=X_train.columns.to_list(),
    class_names=['dead', 'survive']
)
viz.save('./output/tree_titanic.svg')
viz
```

## 多クラス分類

```python
iris = load_iris()

# 特徴量と目的変数の分離
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = pd.Series(iris.target, name='class')

print(X.shape, y.shape)
display(X.head())
display(y.head())

# train/test分割 (train:test = 7:3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  shuffle=True, random_state=100)
print('学習データ : ', X_train.shape, y_train.shape)
print('テストデータ : ', X_test.shape, y_test.shape)
```

```python
# モデル学習
clf = tree.DecisionTreeClassifier(max_depth=5, min_samples_split=10, random_state=100)
clf = clf.fit(X_train, y_train)

# 予測値算出
train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test) 
print(train_pred.shape, test_pred.shape)

# 評価指標算出
calc_clf_metrics(y_train, train_pred, title='Metrics (train)')
calc_clf_metrics(y_test, test_pred, title='Metrics (test)')

# Confusion matrix作成
class_names = iris.target_names
print('\n')
plot_confusion_matrix(y_train, train_pred, normalize=True, classes=class_names)
print('\n')
plot_confusion_matrix(y_test, test_pred, normalize=True, classes=class_names)
```

```python
# 決定木の描画
viz = dtreeviz(
    clf,
    X_train,
    y_train,
    target_name='class',
    feature_names=X_train.columns.to_list(),
    class_names=[str(i) for i in iris.target_names]
)
viz.save('./output/tree_iris.svg')
viz
```

## 回帰

```python
boston = load_boston()

# 特徴量と目的変数の分離
X = pd.DataFrame(boston.data, columns=boston.feature_names)
y = pd.Series(boston.target, name='target')

print(X.shape, y.shape)
display(X.head())
display(y.head())

# train/test分割 (train:test = 7:3)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  shuffle=True, random_state=100)
print('学習データ : ', X_train.shape, y_train.shape)
print('テストデータ : ', X_test.shape, y_test.shape)
```

```python
# モデル学習
model = tree.DecisionTreeRegressor(criterion='mse', max_depth=5, random_state=100)
model = model.fit(X_train, y_train)

# 予測値算出
train_pred = model.predict(X_train)
test_pred = model.predict(X_test) 
print(train_pred.shape, test_pred.shape)

# 評価指標算出
calc_mse(y_train, train_pred, title='MSE (train)]')
calc_mse(y_test, test_pred, title='MSE (test)]')

# テストデータの予測値分布の描画
fig, ax = plt.subplots(figsize=(8, 8))
sns.histplot(y_test, label='Test', ax=ax, color='black')
sns.histplot(test_pred, label='Test Predicted', ax=ax, color='C1')
ax.legend()
ax.grid()
```

```python
# 決定木の描画
viz = dtreeviz(
    model,
    X_train,
    y_train,
    target_name='house price',
    feature_names=X_train.columns.to_list()
)
viz.save('./output/tree_boston.svg')
viz
```
